```python
#import libraries
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px

from scipy.stats import linregress
```

```python
#pull in csv file

data_df = data_df = pd.read_csv("../Data Sets/all_cancer_and_pesticides.csv")
data_df.head()
```

```python
#formatting
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', '{:.9f}'.format)

# Define the pesticide columns
pesticide_columns = ['Insecticide Sum', 'Herbicide Sum', 'Fungicide Sum']

# Define the cancer type columns
cancer_columns = [
    'Cancer Incidence Rate Per 100k',
    'Breast Cancer Incidence Rate Per 100k',
    'Leukemia Incidence Rate Per 100k',
    'Lung Cancer Incidence Rate Per 100k',
    'Non-Hodgkin Lymphoma Incidence Rate Per 100k',
    'Prostate Cancer Incidence Rate Per 100k'
]

# Filter out Union County and Palo Alto County
filtered_df = data_df[
    ~data_df['County'].isin(['Union County, Florida(6)', 'Palo Alto County, Iowa(7)'])
]

# Initialize a list to store the summary statistics
summary_statistics = []

# Calculate the slope and R score for each combination of pesticide and cancer type
for pesticide in pesticide_columns:
    for cancer in cancer_columns:
        slope, intercept, r_value, p_value, std_err = linregress(filtered_df[pesticide], filtered_df[cancer])
        summary_statistics.append({'Pesticide': pesticide, 'Cancer Type': cancer, 'Slope': round(slope, 9), 'R score': r_value})

# Create a DataFrame to hold the summary statistics
summary_df = pd.DataFrame(summary_statistics)

# Display the summary statistics table
summary_df.sort_values(by='Cancer Type')
```
