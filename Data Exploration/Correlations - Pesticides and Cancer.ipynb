**Pesticide data from USDA, 2019 units are Kilograms/year. Cancer Data from 2016-2020 (5 Year average)**

```python
#import libraries
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px

from scipy.stats import linregress
```

```python
#pull in csv files. 
data_df = pd.read_csv("../Data Sets/cancer_and_pesticides.csv")
#data_df.head()

detailed_cancer_df = data_df = pd.read_csv("../Data Sets/all_cancer_and_pesticides.csv")
#detailed_cancer_df.head()
```

```python
#Union County Florida seems to be an outlier with low pesticide use rates and very high cancer rates.
#Let's remove it to see if it is effecting correlation calculations 

filtered_data_df = data_df[data_df['County'] != 'Union County, Florida(6)']
```

# Overall Cancer Incidence Rate and Pesticides

## Insecticide

```python
#create R score for cancer incidence rate per 100k and insecticide sum

i_r_score = data_df['Cancer Incidence Rate Per 100k'].corr(data_df['Insecticide Sum'])
i_filtered_r_score = filtered_data_df['Cancer Incidence Rate Per 100k'].corr(filtered_data_df['Insecticide Sum'])
print("R score for Cancer Incidence Rate Per 100k and Insecticide Sum:", i_r_score)
print("Filtered R score, Cancer Incidence Rate/100k & Insecticide Sum:", i_filtered_r_score)
```

**Based on the above, we are proceeding with the filtered data, removing Union County, Florida** 

```python
#Create a graph with line of best fit & data points plotting cancer incidence rate per 100k and insecticide sum

# Set the style of the seaborn plot
sns.set_style("whitegrid")

# Create a scatter plot with a line of best fit
plt.figure(figsize=(10, 6))
sns.regplot(x='Insecticide Sum', y='Cancer Incidence Rate Per 100k', data=filtered_data_df, scatter_kws={'alpha':0.5},
            line_kws={'color': 'darkorange'})

# Set the title and labels for the plot
plt.title('Cancer Incidence Rate Per 100k vs. Insecticide Sum')
plt.xlabel('Insecticide Sum')
plt.ylabel('Cancer Incidence Rate Per 100k')

# Show the plot
#plt.show()
```

```python
#Calculate the line of best fit for  Insecticide Sum and Cancer Incidence Rate
slope, intercept, r_value, p_value, std_err = linregress(filtered_data_df['Insecticide Sum'],
                                                         filtered_data_df['Cancer Incidence Rate Per 100k'])

# Format the equation as a string
equation = f'y = {slope:.7f}x + {intercept:.2f}'
print("Equation of the line of best fit:", equation)
```

**I left the these plots the same to allow for interactive exploration by our team**

```python
#Create an interactive plot to explore the details of the data points.
fig = px.scatter(data_df, x='Insecticide Sum', y='Cancer Incidence Rate Per 100k',
                 hover_data=['County', 'Insecticide Sum', 'Cancer Incidence Rate Per 100k'],
                 title='Cancer Incidence Rate Per 100k vs. Insecticide Sum')

# Show the plot
#fig.show()
```
